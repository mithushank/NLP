```python
from langchain.agents import Tool,initialize_agent
from langchain.llms import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
openai_api_key = os.getenv("OPENAI_API_KEY")
print(openai_api_key)

```

```python
from langchain_community.tools.ddg_search.tool import DuckDuckGoSearchTool
from langchain_community.tools.wikipedia.tool import WikipediaAPIWrapper



search_tools = Tool(
    name="DuckDuckGo Search",
    func= DuckDuckGoSearchTool().run,
    description="Search the web using DuckDuckGo"
)

llm = OpenAI(temperature=0.5)

tools = Tool(
    name="Search",
    func=search_tools.run,
    description="Use this tool to search the web"
)

agent = initialize_agent(tools=tools, llm=llm, agent="zero-shot-react-description", verbose=True)
response = agent.run("Search for the capital of France")
print(response)
```

```python
from langchain.agents import initialize_agent, Tool
from langchain_community.llms import OpenAI
from langchain_community.tools.ddg_search.tool import DuckDuckGoSearchTool

# Initialize DuckDuckGo Search Tool
ddg_search_tool = DuckDuckGoSearchTool()

# Define the tool
tools = [
    Tool(
        name="DuckDuckGo Search",
        func=ddg_search_tool.run,  # Use the `run` method of the DuckDuckGoSearchTool
        description="Search the web using DuckDuckGo"
    )
]

# Initialize OpenAI LLM
llm = OpenAI(temperature=0.5)

# Initialize the agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent="zero-shot-react-description",  # Correct agent type
    verbose=True
)

# Run the agent
response = agent.run("What is the capital of France?")
print(response)
```

```python
from langchain.agents import initialize_agent, Tool
from langchain_community.llms import Ollama  # Use Ollama for Llama 3.2
from langchain_community.tools.ddg_search.tool import DuckDuckGoSearchTool
from langchain_community.tools.wikipedia.tool import WikipediaAPIWrapper
from langchain_community.tools.arxiv.tool import ArxivQueryRun
from langchain_community.tools.file_management import ReadFileTool

from langchain_community.utilities.sql_database import SQLDatabase
from langchain_community.agent_toolkits.sql.toolkit import SQLDatabaseToolkit



# Initialize DuckDuckGo Search Tool
ddg_search_tool = DuckDuckGoSearchTool()


search_tool = Tool(
    name="DuckDuckGo Search",
    func=ddg_search_tool.run,
    description="Search the web using DuckDuckGo"
)

wikipedia_tool = Tool(
    name="Wikipedia Search",
    func=WikipediaAPIWrapper().run,
    description="Search in Wikipedia articles to findout the context"
)

read_file_tool = Tool(
    name = "Read File",
    func = ReadFileTool().run,
    description="use this tool to read the content in the file"
)

arxiv_tool = Tool(
    name="Arxiv Search",	
    func=ArxivQueryRun().run,
    description="Search for research papers on Arxiv"
)


# Define the tool
tools = [
    search_tool,
    wikipedia_tool,
    read_file_tool,
    arxiv_tool
]

# Initialize Llama 3.2 using Ollama
llm = Ollama(model="llama3.2")  # Use the Llama 3.2 model

# Initialize the agent
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent="zero-shot-react-description",  # Correct agent type
    verbose=True
)

# Run the agent
response = agent.run("What is the capital of France?")
print(response)
```
